```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('order.csv')
df
```

```python
scaler = MinMaxScaler()
```

```python
df['scaled_order'] = scaler.fit_transform(df[['2023_ORDER']])
df['scaled_score'] = scaler.fit_transform(df[['SCORE']])
```

```python
df
```

```python
df['sum_scaled_columns'] = df['scaled_order'] + df['scaled_score'] * -1
```

```python
df
```

```python
df_sorted = df.sort_values(by='sum_scaled_columns', ascending=False)
df_sorted
```

```python
df_sorted.to_csv('order_new.csv')
```

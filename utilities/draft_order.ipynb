```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('order.csv')
df
```

```python
scaler = MinMaxScaler()
```

```python
df['SCALED_ORDER'] = scaler.fit_transform(df[['2023_ORDER']]).round(3)
df['SCALED_SCORE'] = scaler.fit_transform(df[['SCORE']]).round(3)
```

```python
df
```

```python
df['TOTAL'] = df['SCALED_ORDER'] + df['SCALED_SCORE'] * -1
```

```python
df
```

```python
df_sorted = df.sort_values(by='TOTAL', ascending=False)
df_sorted
```

```python
df_sorted.to_csv('order_new.csv')
```
